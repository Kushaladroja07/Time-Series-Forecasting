```python
#time resampling

df_tesla.head()
```

```python
df_tesla['Date'] = pd.to_datetime(df_tesla['Date'])
```

```python
df_tesla.set_index('Date', inplace=True)
```

```python
df_tesla.resample('YE').min()
```

```python
df_tesla.resample('YE').max()['Open'].plot()
```

```python
df_tesla.resample('QS').max()['Open'].plot()
```

```python
df_tesla.resample('BQS').max()['Open'].plot()
```

```python
df_tesla['Open'].resample('QS').mean().plot( figsize=(15,6))
```

```python
import matplotlib.pyplot as plt

quarterly = df_tesla['Open'].resample('QS').mean()

quarterly.index = quarterly.index.strftime('%Y-%m')

quarterly.plot(kind='bar', figsize=(15,6))

plt.show()
```
